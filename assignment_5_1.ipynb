```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame({
    "price":[110,105,115,120,110,130,150,100,105]
})
```

```python
df
```

# Normalization:

### Mathematical Eqn:
  - `Xnew = (Xi - min(X)) / (max(x) - min(x))`

```python
min_price = df['price'].min()
min_price
```

```python
max_price = df['price'].max()
max_price
```

```python
df['price_normal_manual'] = (df['price'] - min_price ) / (max_price - min_price)
```

```python
df
```

### SKlearn method

```python
from sklearn.preprocessing import MinMaxScaler
mmx = MinMaxScaler() #feature_range=(2,5)
```

```python
df['price_normal_sklearn'] = mmx.fit_transform(df[['price']])
df
```

# Standardization / standard scaler

### Mathematical equation:
  - `Xnew = (Xi - Xmean ) / standard_deviation`
  - `standard_deviation:
  - image.png

  - There can be 2 standard deviation
    - 1. N
    - 2. N-1

```python
mean_price = df['price'].mean()
mean_price
```

```python
standard_deviation =  df['price'].std()
standard_deviation
```

```python
df['price_std_manual'] = (df['price'] - mean_price) / standard_deviation
df
```

- ### Sklearn

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df['price_std_sklearn'] = scaler.fit_transform(df[['price']])

df
```

```python
standard_deviation_ddof_0 = df['price'].std(ddof=0)
standard_deviation_ddof_0
```

```python
standard_deviation_ddof_1 = df['price'].std(ddof=1)
standard_deviation_ddof_1
```

```python
df['price_std_manual_2'] = (df['price'] - mean_price) / standard_deviation_ddof_0
df
```

# Log Transformation
